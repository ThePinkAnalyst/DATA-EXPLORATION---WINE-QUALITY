# Dealing with outliers

Box plot method used to identify outliers visually and using IQR to select outliers

```python
#Box plot Fixed acidity
df['fixed acidity'].plot(kind='box')
```

```python
#Replace outliers with null values
for x in ['fixed acidity']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
    
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan 
```
